## Remove the outliers

```python
df.info()
```

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
numeric_columns
```

```python
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(16, 12))
axes = axes.flatten()

for i, column in enumerate(numeric_columns):
    sns.histplot(df[column],kde=True, ax=axes[i])
    axes[i].set_title(column)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')
plt.show()
```

```python
def check_outliers(column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lw = q1 - (1.5 * iqr)
    up = q3 + (1.5 * iqr)

    df[column] = df[column].apply(lambda x: lw if x < lw else ( up if x> up else x))
    return lw, up
```

```python
fig ,ax = plt.subplots(nrows=3,ncols=4,figsize=(16,12))
ax = ax.flatten()
for i,column in enumerate(numeric_columns):
     sns.boxplot(df[column],ax=ax[i])
     ax[i].set_title(column)
     ax[i].set_ylabel("")
```

```python
for i in numeric_columns:
     check_outliers(i)
```

```python
fig ,ax = plt.subplots(nrows=3,ncols=4,figsize=(16,12))
ax = ax.flatten()
for i,column in enumerate(numeric_columns):
     sns.boxplot(df[column],ax=ax[i])
     ax[i].set_title(column)
     ax[i].set_ylabel("")
```
