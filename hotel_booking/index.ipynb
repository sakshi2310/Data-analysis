### What is the percentage of booking for each year? 

```python
# What is the percentage of booking for each year?
df["arrival_date_year"].unique()
per_2015 = (
    df[df["arrival_date_year"] == 2015]["arrival_date_year"].count() / df.shape[0] * 100
)

per_2016 = (
    df[df["arrival_date_year"] == 2016]["arrival_date_year"].count() / df.shape[0] * 100
)

per_2017 = (
    df[df["arrival_date_year"] == 2017]["arrival_date_year"].count() / df.shape[0] * 100
)
```

```python
comare_year = pd.DataFrame(
    [per_2015, per_2016, per_2017], columns=["percentage"], index=[2015, 2016, 2017]
).reset_index(names="Year")
comare_year
```

```python
plt.bar(comare_year["Year"], comare_year["percentage"])
plt.xticks([2015, 2016, 2017])
```
